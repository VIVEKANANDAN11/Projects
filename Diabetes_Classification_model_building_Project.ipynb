```python
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(x)
X
```

***Train Test Split***

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=119)
```

```python
y_test
```

***Classification Methodologies***

***KNN[K Near Neighbour] Method With Evaluation Metrices***

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=7,p=2 )
classifier.fit(x_train, y_train)
```

```python
y_pred_knn=classifier.predict(x_test)
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_pred_knn)
print('Confusion',cm)
recall = recall_score(y_test,y_pred_knn)
print('Recall',recall)
accuracy = accuracy_score(y_test,y_pred_knn)
print('Accuracy',accuracy)
precision = precision_score(y_test,y_pred_knn)
print('Precision',precision)
F1score = f1_score(y_test,y_pred_knn)
print('F1 score',F1score)
```

***SVM Methodology with Evaluation Metrices***

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel = 'linear', random_state = 0)
classifier2.fit(x_train, y_train)
```

```python
y_pred_svm= classifier2.predict(x_test)
y_pred_svm
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_pred_svm)
print('Confusion',cm)
recall = recall_score(y_test,y_pred_svm)
print('Recall',recall)
accuracy = accuracy_score(y_test,y_pred_svm)
print('Accuracy',accuracy)
precision = precision_score(y_test,y_pred_svm)
print('Precision',precision)
F1score = f1_score(y_test,y_pred_svm)
print('F1 score',F1score)
```

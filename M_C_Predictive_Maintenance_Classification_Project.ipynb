```python
df1=pd.read_csv('predictive_maintenance.csv',index_col="UDI")
df1
```

```python
x=df1.iloc[:,[0,1,2,3,4,5,6,7]]
x
```

```python
df1.Target.unique()
```

```python
y=df1[['Failure Type']]
y
```

```python
df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x['Product ID']=le.fit_transform(x['Product ID'])
x['Type']=le.fit_transform(x['Type'])
x['Air temperature [K]']=le.fit_transform(x['Air temperature [K]'])
x['Process temperature [K]']=le.fit_transform(x['Process temperature [K]'])
x['Torque [Nm]']=le.fit_transform(x['Torque [Nm]'])
x
```

```python
y['Failure Type']=le.fit_transform(y['Failure Type'])
y.values.ravel()
```

```python
x.info()
```

```python
Heatmap=x.corr()
sns.heatmap(Heatmap,annot=True,cmap='coolwarm')
#mlp.show()
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_std=std.fit_transform(x)
x_std
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x_std,y,test_size=0.2,random_state=1117)  # 80:20 , 70:30, # parado principle
x_train, x_test, y_train, y_test
```

```python
def conf_matrixs(y_test,model,x_test,cmap,normalize=None,plot=True,encoded_labels=True):
    y_pred = model.predict(x_test)
    conf_mat = confusion_matrix(y_test,y_pred,normalize=None,labels=[0,1,2,3,4,5])
    ax = sns.heatmap(conf_mat, cmap=cmap, square=True, cbar=False, annot=True, fmt='g')
    ax.set_title('confusion_matrics')
    ax.set_xlabel('prediction')
    ax.set_ylabel('Actual')
    return conf_mat
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN= KNeighborsClassifier(n_neighbors=5,p=2 )
KNN.fit(x_train,y_train)
```

```python
y_predict_KNN =KNN.predict(x_test)
y_predict_KNN
```

```python
conf_matrixs(y_test,KNN,x_test,cmap='inferno',normalize=None,plot=True,encoded_labels=True)
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_predict_KNN)
print('Confusion matrix','\n',cm)
accuracy = accuracy_score(y_test,y_predict_KNN)
print('Accuracy',accuracy)
recall = recall_score(y_test,y_predict_KNN,labels=[0,1,2,3,4,5],average='macro')
print('Recall',recall)
precision = precision_score(y_test,y_predict_KNN,labels=[0,1,2,3,4,5],average='macro',zero_division=1)
print('Precision',precision)
F1score = f1_score(y_test,y_predict_KNN,labels=[0,1,2,3,4,5],average='macro')
print('F1 score',F1score)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_predict_KNN,labels=[1,2,3,4,5,6],zero_division=1,target_names=['0','1','2','3','4','5'])
```
